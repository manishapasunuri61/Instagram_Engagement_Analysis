## **Instagram Improving Quality and Retention**

### **Business Context**

##### Instagram is one of the world's largest social platforms, used daily for entertainment, discovery, and communication across formats like Feed (photos), Reels, Stories, Explore, and DMs. <br>
##### As Instagram scales, leadership's focus is shifting from “more time spent” to Engagement Quality — making sure that user activity is healthy, consistent, and sustainable, rather than driven by unhealthy patterns (doom-scrolling), notification spam, or low-quality interaction (passive views with no meaningful actions). <br>
##### **Poor engagement quality creates:** <br> lower long-term retention (users churn after heavy short-term usage),<br> weaker creator ecosystem (less meaningful engagement),<br> higher trust/safety risk (account security issues, privacy distrust),<br> weaker monetization (premium users don't stick, ad engagement declines). <br>
##### Instagram believes the answers lie in its user lifestyle + usage behavior dataset, capturing engagement patterns, content preferences, notification behavior, and account settings.

### **Objectives** <br>

##### To explore the dataset and uncover insights to improve Engagement Quality Efficiency (EQE) and reduce churn risk. <br>
##### **Deconstruct EQE across key dimensions:**
demographics: age, gender, etc.
<br> platform behavior: sessions_per_day, etc.
<br> content: content_type_preference, preferred_content_theme
<br> account trust: privacy_setting_level, etc.
<br>
##### **Identify which combinations drive:**
high EQE (healthy engagement),
<br> low EQE (passive doom-scroll patterns),
<br> dormancy risk (low sessions + low notif response + old last_login_date),
etc.
<br>
##### **Study lifestyle links:**
Check if health/lifestyle correlate with engagement quality:
<br> sleep_hours_per_night, exercise_hours_per_week, perceived_stress_score, etc.
<br> and see association with key metrics
<br>
##### **Surface actionable insights**
**Recommend actions to:**
<br> improve feed ranking for meaningful engagement,
<br> reduce passive scrolling loops,
<br> personalize notifications better,
<br> improve retention for at-risk cohorts,
<br> increase premium adoption (only if it improves EQE and retention).

## **Calculating Key Metrics**

#### **Calculating Attention (A)** Time Spent

```python
time_cols = ['time_on_feed_per_day', 'time_on_reels_per_day', 'time_on_explore_per_day', 'time_on_messages_per_day']

insta_usage_df['A'] = insta_usage_df[time_cols].sum(axis = 1)
insta_usage_df[['user_id'] + time_cols + ['A']].head()
```

#### **Calculating Meaningful Engagement (ME)**

Here ME is a weighted score which measures user activity. <br>
It is a function of (likes_given_per_day, comments_written_per_day, dms_sent_per_week, posts_created_per_week, shares/saves)

**So here we give weights according to the efforts of the activity:** <br>
like = low effort , weight = 1 <br>
comment = medium effort, weight = 3 <br>
DM = stronger intent, weight = 4 <br>
post =  strongest action, weight = 5 <br>
shares/saves = this is strong engagement action for analysis but column does not exist in dataset

and also **(dms and post)** are converted to daily action value by **dividing the value by 7**

```python
insta_usage_df['dms_sent_per_day'] = insta_usage_df['dms_sent_per_week'] / 7
insta_usage_df['posts_created_per_day'] = insta_usage_df['posts_created_per_week'] / 7

#calculating ME with weights
insta_usage_df['ME'] = (1 * insta_usage_df['likes_given_per_day'] + 3 * insta_usage_df['comments_written_per_day'] + 4 * insta_usage_df['dms_sent_per_day'] + 5 * insta_usage_df['posts_created_per_day'])

insta_usage_df[['user_id', 'likes_given_per_day', 'comments_written_per_day', 'dms_sent_per_day', 'posts_created_per_day', 'ME']].head()
```

#### **Calculating Engagement Quality Efficiency (EQE)**

```python
# EQE = ME/A
insta_usage_df['EQE'] = np.where(insta_usage_df['A'] > 0, insta_usage_df['ME'] / insta_usage_df['A'], np.nan)

insta_usage_df[['user_id', 'ME', 'A', 'EQE']].head()
```

### **Dormancy risk (low sessions + low notif response + old last_login_date**

```python
# taking most recent date from dataset
ref_date = insta_usage_df['last_login_date'].max()

# calculating days since last login
insta_usage_df['days_since_last_login'] = (ref_date - insta_usage_df['last_login_date']).dt.days

# creating buckets dormancy signals
# 1. sessions bucket - it can be used from combo1 of (low and high eqe) -  sessions_level

# 2. creating notification response bucket (low vs high)
insta_usage_df['notif_response_level'] = pd.qcut(insta_usage_df['notification_response_rate'], 2, labels=['low_notif_response', 'high_notif_response'])

# 3. last login bucket
insta_usage_df['last_login_bucket'] = pd.cut(insta_usage_df['days_since_last_login'], 2, labels=['recent_login', 'old_login'])

# creating dormancy combination key
insta_usage_df['dormancy_combo_key'] = (insta_usage_df['sessions_level'].astype(str) + " | " + insta_usage_df['notif_response_level'].astype(str) + " | " + insta_usage_df['last_login_bucket'].astype(str))

# adding dormancy risk label
insta_usage_df['dormancy_risk'] = np.where((insta_usage_df['sessions_level'].astype(str) == 'low_sessions') & (insta_usage_df['notif_response_level'].astype(str) == 'low_notif_response') & (insta_usage_df['last_login_bucket'].astype(str) == 'old_login'), 'High_dormancy_risk', 'Low_dormancy_risk')

# dormancy risk vs EQE segment
dormancy_risk_vs_eqe = pd.crosstab(insta_usage_df['dormancy_risk'], insta_usage_df['EQE_segment'], normalize = 'index').round(3).reset_index()

# adding high dormancy domiannt label
dormancy_risk_vs_eqe['dominant_segment'] = np.where(dormancy_risk_vs_eqe['High_EQE'] > dormancy_risk_vs_eqe['Low_EQE'], 'High EQE dominant', 'Low EQE dominant')

dormancy_risk_vs_eqe
```

**Insights:**

1. **High dormancy risk** users are mostly **High_EQE**, it means they engage efficiently but login less often.
2. **Low dormancy risk** users are mostly **Low_EQE**, they login very often but usage is on scrolling rather than meaningful actions.

**Recommendations:**
1. Focus on improving EQE for **low dormancy risk users** by promoting **(comments + DMs + saves/shares)** instead of just spending time.
